# Bias, Ca, and the eq. potential.

```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
output_notebook()

import numpy as np

from fakespikes.util import kappa
from fakespikes.util import fano

from resistingrhythm.neurons import HHH
```

```python
# ------------------------------------------------
# Of interest
bias_in=150e-9

Ca_target = 0
Ca = 0

# ------------------------------------------------
# Fixed
time = 3
time_step = 1e-4
N = 1

V_e = 0
V_i = -80e-3

tau_e = 5e-3
tau_i = 20e-3
tau_h = 1

w = 2e-6

# ------------------------------------------------
results0 = HHH(
    time,
    np.asarray([]),
    np.asarray([]),
    np.asarray([]), # osc
    np.asarray([]),
    N=N,
    Ca=0,
    Ca_target=0,
    tau_h=tau_h,
    w_in=w,
    tau_in=tau_e,
    V_in=V_e,
    bias_in=0,
    w_osc=w,
    tau_osc=tau_e,
    V_osc=V_e,
    sigma=0,
    time_step=time_step
)

results1 = HHH(
    time,
    np.asarray([]),
    np.asarray([]),
    np.asarray([]), # osc
    np.asarray([]),
    N=N,
    Ca=Ca,
    Ca_target=Ca_target,
    tau_h=tau_h,
    w_in=w,
    tau_in=tau_e,
    V_in=V_e,
    bias_in=bias_in,
    w_osc=w,
    tau_osc=tau_e,
    V_osc=V_e,
    sigma=0,
    time_step=time_step
)
```

```python
p = figure(plot_width=400, plot_height=200)
for n in range(N):
    p.line(x=results0['times'], y=results0['v_m'][n, :], color="grey", alpha=0.5)
    p.line(x=results1['times'], y=results1['v_m'][n, :], color="purple", alpha=0.5)
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'V_m (volts)'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.x_range = Range1d(0, time)
p.y_range = Range1d(-70e-3, -50e-3)
show(p)
```
