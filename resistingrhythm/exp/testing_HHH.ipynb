# Basic tests for the HHH neuron

```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
output_notebook()

import numpy as np

from resistingrhythm.util import burst
from resistingrhythm.util import pulse
from resistingrhythm.util import poisson_impulse
from resistingrhythm.neurons import HHH
```

# Shared params

```python
time = 0.6
N = 1

V_e = 0
V_i = -80e-3

tau_e = 5e-3
tau_i = 20e-3

w = 1e-6
```

# Generate some input....

```python
t = 0.4
stim_number = 10
stim_onset = 0.1
stim_offset = 0.5
stim_rate = 60
dt = 1e-5
seed_stim = 7525
ns_in, ts_in = poisson_impulse(
        t,
        stim_onset,
        stim_offset - stim_onset,
        stim_rate,
        n=stim_number,
        dt=dt,
        seed=seed_stim)
```

# Run HHH

```python
results = HHH(
    time,
    ns_in,
    ts_in,
    np.asarray([]), # osc
    np.asarray([]),
    N=N,
    w_in=w,
    tau_in=tau_e,
    V_in=V_e,
    bias_in=0.0e-9,
    w_osc=w,
    tau_osc=tau_e,
    V_osc=V_e,
    sigma=0
)
```

```python
results.keys()
results['v_m'].shape
```

```python
p = figure(plot_width=400, plot_height=200)
p.circle(ts_in, ns_in, color="grey")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'N'
p.x_range = Range1d(0, time)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```

```python
p = figure(plot_width=400, plot_height=200)
p.circle(results['ts'], results['ns'], color="black")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'N'
p.x_range = Range1d(0, time)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```

```python
p = figure(plot_width=400, plot_height=200)
for n in range(N):
    v = results['g_total'][n, :]
    p.line(x=results['times'], y=v, color="black", alpha=0.5)
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'g_total (volts)'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
# p.y_range = Range1d(0, 1000e-9)
show(p)
```

```python
p = figure(plot_width=400, plot_height=200)
for n in range(N):
    v = results['v_m'][n, :]
    p.line(x=results['times'], y=v, color="black", alpha=0.5)
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'V_m (volts)'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
# p.y_range = Range1d(0, 1000e-9)
show(p)
```
